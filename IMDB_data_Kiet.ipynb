```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load
movies_data_to_load = Path("imdb_movies.csv")

# Read and store into Pandas DataFrames
movies_data = pd.read_csv(movies_data_to_load)
movies_data.head()
```

```python
# Trim the data
movies_data_reduced = movies_data.drop(["overview","crew","orig_title","status"], axis=1)
movies_data_reduced
```

```python
positive_revenue = movies_data_reduced.loc[movies_data_reduced["revenue"]>0]
positive_revenue
```

```python
# Double check this code, it seems to have an issue
# start_date = "01/01/1990"
# end_date = "01/01/2020"
# date_filter_df = positive_revenue.loc[(positive_revenue["date_x"]>=start_date) & (positive_revenue["date_x"]<=end_date)]
# date_filter_df

# Assuming date_x is stored as string
positive_revenue.loc[:, "date_x"] = pd.to_datetime(positive_revenue["date_x"])

start_date = pd.to_datetime("01/01/1990")
end_date = pd.to_datetime("12/31/2020")

date_filter_df = positive_revenue.loc[(positive_revenue["date_x"] >= start_date) & (positive_revenue["date_x"] <= end_date)]
date_filter_df
```

```python
English_movies = date_filter_df.loc[date_filter_df["orig_lang"]== " English"]
English_movies
```

```python
Non_English_movies = date_filter_df.loc[date_filter_df["orig_lang"]!= " English"]
Non_English_movies
```
